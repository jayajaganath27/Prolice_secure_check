```python
create_table_query = """ CREATE TABLE IF NOT EXISTS police_log (
stop_date DATETIME,
stop_time   DATETIME,
country_name  VARCHAR(200),
driver_gender VARCHAR(10),
driver_age     INT,
driver_race VARCHAR(10),
violation_raw  VARCHAR(200),
violation       VARCHAR(200),
search_conducted VARCHAR(200),
search_type      VARCHAR(200),
stop_outcome      VARCHAR(200),
is_arrested       BOOLEAN,
stop_duration      VARCHAR(200),
drugs_related_stop  BOOLEAN,
vehicle_number      VARCHAR(200))"""

mycursor.execute(create_table_query)
print("✅Table created sucessfully")
```

```python
insert_query = """
INSERT INTO police_log (
  stop_date,
  stop_time,
  country_name,
  driver_gender,
  driver_age,
  driver_race,
  violation_raw,
  violation,
  search_conducted,
  search_type,
  stop_outcome,
  is_arrested,
  stop_duration,
  drugs_related_stop,
  vehicle_number
) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
```

```python
for index, row in securecheck.iterrows():
    stop_date = row['stop_date'].strftime('%Y-%m-%d') if pd.notnull(row['stop_date']) else None
    stop_time = row['stop_time'].strftime('%H:%M:%S') if pd.notnull(row['stop_time']) else None
    stop_datetime = f"{stop_date} {stop_time}" if stop_date and stop_time else None

    mycursor.execute(insert_query,(
        stop_date,
        stop_datetime,
        row.get('country_name'),
        row.get('driver_gender'),
        row.get('driver_age'),
        row.get('driver_race'),
        row.get('violation_raw'),
        row.get('violation'),
        row.get('search_conducted'),
        row.get('search_type'),
        row.get('stop_outcome'),
        row.get('is_arrested'),
        row.get('stop_duration'),
        row.get('drugs_related_stop'),
        row.get('vehicle_number')
    ))

mydb.commit()
mycursor.close()
print("✅ All records inserted successfully!")
```
